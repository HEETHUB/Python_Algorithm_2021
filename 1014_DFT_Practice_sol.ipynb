```python
class TreeNode():
    def __init__(self, s: str, k: int):
        self.name = s
        self.ary = k
        self.child = [None]*k
```

```python
class Tree():
    def __init__(self, root: TreeNode) -> None:
        self.root = root
    
    def visit(self, node: TreeNode, depth:int) -> None:
        print("--"*depth, node.name)
        
    def __DFT_preorderHelp(self, curNode, depth:int)->None:
        #base case
        if curNode ==None:
            return
        self.visit(curNode, depth)

        #child
        for childNode in curNode.child :
            self.__DFT_preorderHelp(childNode, depth+1)

        
    def DFT_preorder(self) -> None:
        self.__DFT_preorderHelp(self.root, 0)
```

# Your code for practice (skeleton for recursive implementation)

```python
# Stack-based implementation
from collections import deque
```

```python
class Tree():
    def __init__(self, root: TreeNode) -> None:
        self.root = root
    
    def visit(self, node: TreeNode, depth:int) -> None:
        print("--"*depth, node.name)
        
        
    def DFT_preorder(self):
        curNode = self.root
        depth = 0
        child_idx = 0

        stack = deque([[curNode, depth, child_idx]])

        while stack :
            # Get the last element of the stack
            curNode, depth, child_idx = stack[-1]

            if curNode ==None:
                stack.pop()
            else:
                  #Preoder visit\
                  # (visiting myself before visiting any of my child)
                  if child_idx ==0:
                      self.visit(curNode, depth)

                  #child visit
                  if child_idx < curNode.ary:
                      stack[-1][2] = child_idx+1  # increment child_idx
                      stack.append([curNode.child[child_idx], depth+1, 0])
                  # Finished visiting children (time to pop myself)
                  else:
                      stack.pop()

```

```python
# Create a tree
myTree = Tree(t_root)

# Execute preorder DFT
myTree.DFT_preorder()
```
